```python
dic = dict()

for status in statuses:
    if 'retweeted_status' in status:
        retweet_counts = status['retweet_count']
        retweeted_ids = status['retweeted_status']['id']
        retweeted_users = status['retweeted_status']['user']['screen_name']
        retweet_texts = status['text']
        
        dic[retweeted_ids] = retweet_counts, retweeted_users, retweet_texts  
        retweets = list(dic.values())

print(retweets)
len(retweets)
```
